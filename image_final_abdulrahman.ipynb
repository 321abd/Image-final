```python
# q3
# Path to the image
image_path_p3 = 'pout.tif'

# Read the image in grayscale 
img_p3_original_gray = cv2.imread(image_path_p3, cv2.IMREAD_GRAYSCALE)

# Check if image was loaded successfully
if img_p3_original_gray is None:
    print(f"Error: Could not load image {image_path_p3}")
else:
    # Display the original grayscale image 
    plt.figure(figsize=(8, 6))
    plt.imshow(img_p3_original_gray, cmap='gray')
    plt.title('Original Image (Grayscale)')
    plt.axis('off')
    plt.show()
```

```python
if img_p3_original_gray is not None:
    # Apply log transform: s = c * log(1 + r)
    # First, normalize pixel values to 0-1 range for log function
    # Then apply log, and scale back to 0-255
    # Choose a constant c (e.g., 255 / log(1 + max_pixel_value))
    c_p3 = 255 / np.log(1 + np.max(img_p3_original_gray))
    img_p3_log_transformed = c_p3 * np.log(1 + img_p3_original_gray)
    img_p3_log_transformed = np.array(img_p3_log_transformed, dtype=np.uint8)

    # Display the log transformed image 
    plt.figure(figsize=(8, 6))
    plt.imshow(img_p3_log_transformed, cmap='gray')
    plt.title('Log Transformed Image')
    plt.axis('off')
    plt.show()
```

```python
if img_p3_original_gray is not None:
    # Apply inverse log transform: s = e^(r/c) - 1
    # Normalize original image to 0-1 before inverse log operation
    normalized_original_p3 = img_p3_original_gray / 255.0
    img_p3_inverse_log_original = (np.exp(normalized_original_p3 * np.log(1 + 255) / c_p3) - 1) * 255
    img_p3_inverse_log_original = np.array(img_p3_inverse_log_original, dtype=np.uint8)

    # Display the inverse log transformed original image 
    plt.figure(figsize=(8, 6))
    plt.imshow(img_p3_inverse_log_original, cmap='gray')
    plt.title('Inverse Log Transformed Original Image')
    plt.axis('off')
    plt.show()
    if img_p3_log_transformed is not None:
        # Apply inverse log transform to the log transformed image
        # s = e^(r/c) - 1
        # Normalize log_transformed image to 0-1 before inverse log operation
        normalized_log_transformed_p3 = img_p3_log_transformed / 255.0
        img_p3_inverse_log_of_log = (np.exp(normalized_log_transformed_p3 * np.log(1 + 255) / c_p3) - 1) * 255
        img_p3_inverse_log_of_log = np.array(img_p3_inverse_log_of_log, dtype=np.uint8)
    
        # Display the inverse log transformed image from step 2 
        plt.figure(figsize=(8, 6))
        plt.imshow(img_p3_inverse_log_of_log, cmap='gray')
        plt.title('Inverse Log of Log Transformed Image')
        plt.axis('off')
        plt.show()
```

```python
print("Commentary on Log and Inverse Log Transforms: ")
print("Step 2 (Log Transformed Image): ")
print("The log transform (s = c * log(1 + r)) tends to expand the range of dark pixel values.")
print("This makes details in darker regions of the original image more visible, while compressing the range of brighter pixel values.")
print("For 'pout.tiff', which typically has dark areas, the log transform enhances the contrast in those darker regions, making features that were previously hard to distinguish more apparent. ")

print("\nStep 4 (Inverse Log of Log Transformed Image): ")
print("Applying the inverse log transform to the log-transformed image ideally reconstructs the original image.")
print("Comparing this image to the original 'pout.tiff' (from Step 1), we can observe that they are very similar.")
print("This demonstrates the reversibility of the log transformation, provided the same constant 'c' is used and no information was lost due to data type conversion or clipping during the forward transform.")
```

```python
# q4
# Path to the image
image_path_p4 = 'moon.tif'

# Read the image in grayscale 
img_p4_original_gray = cv2.imread(image_path_p4, cv2.IMREAD_GRAYSCALE)

# Check if image was loaded successfully
if img_p4_original_gray is None:
    print(f"Error: Could not load image {image_path_p4}")
else:
    # Display the original grayscale image 
    plt.figure(figsize=(8, 6))
    plt.imshow(img_p4_original_gray, cmap='gray')
    plt.title('Original Moon Image (Grayscale)')
    plt.axis('off')
    plt.show()
```

```python
if img_p4_original_gray is not None:
    # Convert image to float32 for calculations to avoid clipping
    f = img_p4_original_gray.astype(np.float32)

    # Apply Gaussian blur to get the smoothed image
    # Kernel size can be adjusted; (5,5) is a common choice
    f_smoothed = cv2.GaussianBlur(f, (5, 5), 0)

    # Calculate the mask: mask = original - smoothed
    mask_p4 = f - f_smoothed

    # Define different k values 
    k_values = [0.2, 0.5, 1.0] # You can select the value of k arbitrarily 

    spatial_results = []
    for k in k_values:
        # Apply unsharp masking formula 
        g_spatial = f + k * mask_p4
        # Clip values to 0-255 and convert back to uint8
        g_spatial = np.clip(g_spatial, 0, 255).astype(np.uint8)
        spatial_results.append(g_spatial)

    print("Unsharp Masking in Spatial Domain applied for k values:", k_values)
```

```python
if img_p4_original_gray is not None:
    # Get image dimensions
    rows, cols = img_p4_original_gray.shape

    # Optimal DFT size (for speed)
    nrows = cv2.getOptimalDFTSize(rows)
    ncols = cv2.getOptimalDFTSize(cols)
    nimg = np.zeros((nrows, ncols), dtype=np.float32)
    nimg[:rows, :cols] = img_p4_original_gray.astype(np.float32)

    # 1. Fourier Transform F(u,v) 
    dft = cv2.dft(nimg, flags=cv2.DFT_COMPLEX_OUTPUT)
    dft_shift = np.fft.fftshift(dft)

    # Create a meshgrid for filter creation
    center_r, center_c = nrows // 2, ncols // 2
    x = np.arange(ncols) - center_c
    y = np.arange(nrows) - center_r
    X, Y = np.meshgrid(x, y)
    D = np.sqrt(X**2 + Y**2)

    # Design High-Pass Filter H_HP(u,v) 
    # We'll use a Gaussian High-Pass Filter
    # Select D0 (cut-off frequency) 
    # D0 can be found by trial or analyzing frequency domain 
    # For 'moon.tiff', a moderate D0 helps enhance craters without excessive noise
    D0_p4 = 30 # You can adjust this value based on experimentation 

    H_HP = 1 - np.exp(-(D**2) / (2 * (D0_p4**2))) # Gaussian High-Pass Filter

    # Ensure H_HP has two channels to match dft_shift
    H_HP = cv2.merge([H_HP, H_HP])

    frequency_results = []
    # Use the same k values from step 2 
    for k in k_values:
        # Apply the frequency domain formula 
        G_freq = (1 + k * H_HP) * dft_shift

        # Inverse Fourier Transform 
        f_ishift = np.fft.ifftshift(G_freq)
        img_back = cv2.idft(f_ishift, flags=cv2.DFT_SCALE | cv2.DFT_REAL_OUTPUT)

        # Clip and convert to uint8
        g_frequency = np.clip(img_back[:rows, :cols], 0, 255).astype(np.uint8)
        frequency_results.append(g_frequency)

    print(f"Unsharp Masking in Frequency Domain applied for k values: {k_values} with D0 = {D0_p4}")
    print("Note: There is no ideal D0 and k, so results may vary.")
    
    if spatial_results and frequency_results:
        fig, axes = plt.subplots(len(k_values), 2, figsize=(15, 5 * len(k_values)))
        fig.suptitle('Unsharp Masking: Spatial vs. Frequency Domain', fontsize=16)
    
        for i, k in enumerate(k_values):
            ax1 = axes[i, 0] if len(k_values) > 1 else axes[0]
            ax2 = axes[i, 1] if len(k_values) > 1 else axes[1]
    
            ax1.imshow(spatial_results[i], cmap='gray')
            ax1.set_title(f'Spatial Domain (k={k})')
            ax1.axis('off')
    
            ax2.imshow(frequency_results[i], cmap='gray')
            ax2.set_title(f'Frequency Domain (k={k})')
            ax2.axis('off')
        plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent title overlap
        plt.show()
    
        print("Example display format: Fig-1 (k=0.2), Fig-2 (k=0.5), Fig-3 (k=1) ")
```

```python
# q5
# Path to the image
image_path_p5 = 'pcb.tif'

# Read the image in grayscale
img_p5_original_gray = cv2.imread(image_path_p5, cv2.IMREAD_GRAYSCALE)

# Check if image was loaded successfully
if img_p5_original_gray is None:
    print(f"Error: Could not load image {image_path_p5}")
else:
    # Display the original grayscale image
    plt.figure(figsize=(8, 6))
    plt.imshow(img_p5_original_gray, cmap='gray')
    plt.title('Original PCB Image (Grayscale)')
    plt.axis('off')
    plt.show()
```

```python
if img_p5_original_gray is not None:
    # Visual Inspection:
    print("Visual Inspection for Noise Type:")
    print("Upon visual inspection of 'pcb.tiff', scattered bright and dark pixels are prominent across the image.")
    print("These sharp, isolated pixel changes are characteristic features of salt-and-pepper noise.")

    # Statistical Tool: Histogram
    plt.figure(figsize=(8, 6))
    plt.hist(img_p5_original_gray.ravel(), bins=256, range=(0, 256), color='gray')
    plt.title('Histogram of Original PCB Image')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    plt.show()

    print("\nAnalysis of Histogram:")
    print("The histogram shows distinct spikes at both ends of the intensity spectrum (around 0 and 255).")
    print("This confirms the presence of salt-and-pepper noise, where 'salt' pixels are maximum intensity (255) and 'pepper' pixels are minimum intensity (0).")

    print("\nConclusion: The image 'pcb.tiff' is primarily corrupted by **salt-and-pepper noise**.")
```

```python
if img_p5_original_gray is not None:
    # Salt-and-pepper noise is best removed by non-linear filters like the Median filter.
    # A Median filter replaces each pixel's value with the median of its neighbors,
    # which is very effective at removing outliers (the salt and pepper) without
    # excessively blurring edges.

    # Apply Median filter
    # Kernel size (e.g., 3x3, 5x5). An odd number is required.
    # A larger kernel will remove more noise but might blur details more.
    # Starting with 3x3 or 5x5 is common.
    filter_size_p5 = 5 # Experiment with 3 or 5
    img_p5_denoised = cv2.medianBlur(img_p5_original_gray, filter_size_p5)

    # Display the denoised image
    plt.figure(figsize=(8, 6))
    plt.imshow(img_p5_denoised, cmap='gray')
    plt.title(f'Denoised Image (Median Filter, Kernel Size {filter_size_p5}x{filter_size_p5})')
    plt.axis('off')
    plt.show()
```

```python
if img_p6_original_gray is not None:
    print("Problem Commentary for 'pollen.tiff': ")

    # Visual Proof:
    print("\nVisual Proof: ")
    print("Visually, the 'pollen.tiff' image appears to have very low contrast. The pollen grains are barely distinguishable from the background, and the overall image looks 'flat' or 'washed out'. The intensity values seem concentrated in a narrow range.")

    # Statistical/Numerical Proofs:
    min_intensity = np.min(img_p6_original_gray)
    max_intensity = np.max(img_p6_original_gray)
    mean_intensity = np.mean(img_p6_original_gray)
    std_dev_intensity = np.std(img_p6_original_gray)

    print(f"\nNumerical/Statistical Proofs: ")
    print(f"Minimum Pixel Intensity: {min_intensity} ")
    print(f"Maximum Pixel Intensity: {max_intensity} ")
    print(f"Mean Pixel Intensity: {mean_intensity:.2f} ")
    print(f"Standard Deviation of Pixel Intensities: {std_dev_intensity:.2f} ")
    print("A low standard deviation value (e.g., around 20-30 for an 8-bit image) indicates that pixel intensities are clustered closely around the mean, signifying low contrast. The observed range (Max - Min) is also relatively small compared to the full 0-255 range.")

    # Histogram Proof:
    plt.figure(figsize=(8, 6))
    plt.hist(img_p6_original_gray.ravel(), bins=256, range=(0, 256), color='gray')
    plt.title('Histogram of Original Pollen Image')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    plt.show()

    print("\nHistogram Proof: ")
    print("The histogram of 'pollen.tiff' is concentrated in a very narrow band of pixel intensities, typically around the mid-range. It does not span the full 0-255 range. This narrow distribution is a definitive visual proof of low contrast, as it indicates a lack of diverse intensity values needed to represent a wide range of brightness and darkness.")

    print("\nConclusion: The primary problem with 'pollen.tiff' is **low contrast**, which obscures details and makes the image appear dull. ")
```

```python
if img_p6_original_gray is not None:
    print("\nApplying Two Independent Approaches to Solve Low Contrast:")

    # Approach 1: Histogram Equalization
    img_p6_equalized = cv2.equalizeHist(img_p6_original_gray)

    # Approach 2: Adaptive Histogram Equalization (CLAHE)
    # Create a CLAHE object
    # clipLimit: Threshold for contrast limiting. Helps prevent noise amplification.
    # tileGridSize: Size of the grid for histogram equalization.
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    img_p6_clahe = clahe.apply(img_p6_original_gray)

    solution_1_img = img_p6_equalized
    solution_2_img = img_p6_clahe
```

```python
if 'solution_1_img' in locals() and 'solution_2_img' in locals():
    # Display results side-by-side for comparison
    fig, axes = plt.subplots(1, 2, figsize=(15, 7))
    fig.suptitle('Pollen Image Enhancement Solutions', fontsize=16)

    # Display Solution 1: Histogram Equalization
    axes[0].imshow(solution_1_img, cmap='gray')
    axes[0].set_title('Solution 1: Histogram Equalization')
    axes[0].axis('off')

    # Display Solution 2: CLAHE
    axes[1].imshow(solution_2_img, cmap='gray')
    axes[1].set_title('Solution 2: Adaptive Histogram Equalization (CLAHE)')
    axes[1].axis('off')

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```
